```python
import requests
import pandas as pd
from datetime import datetime
```

```python
#CoinStats API endpoint and headers
url = "https://api.coinstats.app/public/v1/coins"
headers = {"User-Agent": "Mozilla/5.0"}

#Retrieve the top 20 cryptocurrencies from the API
response = requests.get(url, headers=headers)
data = response.json()["coins"][:20]
data
```

```python
# Extract the cryptocurrency symbols and prices
symbols = [coin["symbol"] for coin in data]
prices = [coin["price"] for coin in data]
```

```python
# Add a timestamp column to the data
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
timestamps = [timestamp] * len(symbols)
```

```python
# Create a pandas DataFrame with the data
df = pd.DataFrame({"symbol": symbols, "price": prices, "timestamp": timestamps})
```

```python
# Print the DataFrame
print(df)
```

```python
id = [coin["id"] for coin in data]
name = [coin["name"] for coin in data]
volume = [coin["volume"] for coin in data]
marketCap = [coin["marketCap"] for coin in data]
websiteUrl = [coin["websiteUrl"] for coin in data]
#contract_Address = [coin["contractAddress"] for coin in data]
priceBtc = [coin["priceBtc"] for coin in data]
rank = [coin["rank"] for coin in data]
priceBtc = [coin["priceBtc"] for coin in data]
```
